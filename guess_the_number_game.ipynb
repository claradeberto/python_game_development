This is an interactive Python game where the computer generates a random number and the player must guess it within a limited number of attempts.

Features:
- Player name personalisation
- Difficulty levels
- Scoring per round
- Session score accumulated across rounds
- Error handling with try/except
- Game statistics tracking


```python
# Import Python module to generate random numbers within a range
import random
```

```python
# Function definitions
def get_player_name():
    """
    Asks the user for their name.
    Only accepts alphabetic characters.
    Capitalises name.
    """

    while True:

        name = input("Enter your name: ").strip()

        if name.isalpha():
            return name.capitalize()

        else:
            print("\nInvalid name. Please use letters only.\n")


def show_instructions(player_name):
    """
    Display the rules of the game to the player.
    
    player_name (str): Name of the player for personalised messages
    """

    print(f"\nHello {player_name}! Welcome to the Guess the Number Game!")
    print("\n" + "-" * 50)
    print("GAME INSTRUCTIONS".center(50))
    print("-" * 50)
    print("The computer will choose a secret number.")
    print("Your goal is to guess the number before you run out of attempts.")

    print("After each guess you will receive feedback:")
    print("  - 'Too High': your guess is larger than the secret number")
    print("  - 'Too Low': your guess is smaller than the secret number")

    print("\nSCORING SYSTEM")
    print("Your score depends on how many attempts there are left when you win (if you win).")
    print("You earn points every time you win a round.")
    print("For each round you play, points are added to your session score.")

    print("\nScore Formula:")
    print("Score = (Remaining Attempts + 1) × Difficulty Multiplier")


    print("\nMultipliers for Each Difficulty Level:")
    print("Easy: x1")
    print("Medium: x2")
    print("Hard: x4")

    print("\nExample 1:")
    print("If you win on the Easy difficulty with 2 attempts left, then Score = (2+1) × 1 = 3")

    print("\nExample 2:")
    print("If you win on the Hard difficulty with 0 attempts left, then Score = (0+1) × 4 = 4")
    print( "-" * 50 + "\n")


def choose_difficulty():
    """
    Asks the player to select a difficulty level.
    Returns:
    max_number (int): highest possible number within the range
    max_attempts (int): number of attempts allowed
    difficulty_multiplier (int): multiplier used for scoring
    """

    while True:

        print("Choose a difficulty level:\n")
        print("1 = Easy")
        print("   Number range: 1–50")
        print("   Attempts allowed: 15")
        print("   Score multiplier: x1\n")

        print("2 = Medium")
        print("   Number range: 1–100")
        print("   Attempts allowed: 12")
        print("   Score multiplier: x2\n")

        print("3 = Hard")
        print("   Number range: 1–500")
        print("   Attempts allowed: 10")
        print("   Score multiplier: x4\n")

        choice = input("Enter 1, 2, or 3: ")

        if choice == "1":
            return 50, 15, 1

        elif choice == "2":
            return 100, 12, 2

        elif choice == "3":
            return 500, 10, 4

        else:
            print("\nInvalid selection. Please choose 1, 2, or 3.\n")


def get_valid_guess(max_number):
    """
    Asks the player for a guess and validates the input.

    Uses try/except to catch non-numeric inputs.
    """

    while True:

        try:
            guess = int(input("Enter your guess: "))

            if guess < 1 or guess > max_number:
                print(f"Your guess is not within the range. It must be between 1 and {max_number}.")

            else:
                return guess

        except ValueError:
            print("Invalid input. Please enter a NUMBER within the range.\n")


def calculate_score(attempts_used, max_attempts, multiplier):
    """
    Calculates the player's score.

    Score is based on remaining attempts and difficulty multiplier.
    
    Easy difficulty: remaining attempts multiplied by 1.
    Medium difficulty: remaining attempts multiplied by 2.
    Hard difficulty: remaining attempts multiplied by 4.
    """

    remaining_attempts = max_attempts - attempts_used
    score = (remaining_attempts + 1) * multiplier

    return score


def play_game(player_name, stats):
    """
    Runs a single round of the game.

    Updates statistics depending on win or loss.
    """

    # Select difficulty
    max_number, max_attempts, multiplier = choose_difficulty()

    # Generate a random secret number within the selected range
    secret_number = random.randint(1, max_number)

    # Starting number of attempts used
    attempts_used = 0

    print(f"\nLet's begin, {player_name}!")
    print(f"I have generated a number between 1 and {max_number}.")
    print("Try to guess it!\n")

    # Main guessing loop
    while attempts_used < max_attempts:

        guess = get_valid_guess(max_number)

        attempts_used += 1

        # Check result
        if guess == secret_number:

            print(f"\nThat is the correct number {player_name}!")
            print(f"You guessed the number in {attempts_used} attempts.")

            score = calculate_score(attempts_used, max_attempts, multiplier)

            # Update statistics
            stats["session_score"] += score

            print(f"Round score: {score}")
            print(f"Session score: {stats['session_score']}\n")

            stats["games_won"] += 1
            stats["total_attempts"] += attempts_used
            stats["games_played"] += 1

            if score > stats["best_score"]:
                stats["best_score"] = score

            return

        elif guess > secret_number:
            print(f"\n{player_name}, your guess is too HIGH!")

        else:
            print(f"\n{player_name}, your guess is too LOW!")

        # Remaining attempts
        remaining = max_attempts - attempts_used
        
        print(f"Attempts remaining: {remaining}")
        print() # blank line before the next guess

    # If player runs out of attempts
    print(f"\nGame Over {player_name}!")
    print(f"The secret number was: {secret_number}\n")

    stats["games_lost"] += 1
    stats["total_attempts"] += attempts_used
    stats["games_played"] += 1

def ask_to_play_again():
    """
    Asks the player whether they want to play again.
    Only accepts yes/no answers
    The answers are case insensitive.
    Asks for confirmation if the user chooses 'no'.
    Invalid confirmation inputs repeat the confirmation question.
    """

    while True:

        answer = input("Would you like to play again? (yes/no): ").lower()

        if answer == "yes":
            return True

        elif answer == "no":

            while True:

                confirm = input("Are you sure you want to stop playing? (yes/no): ").lower()

                if confirm == "yes":
                    return False

                elif confirm == "no":
                    break

                else:
                    print("\nPlease answer yes or no.\n")

        else:
            print("\nI do not understand whether you want to play again or not.")
            print("Please respond yes or no.\n")


def show_statistics(stats):
    """
    Displays the overall game statistics after the player stops playing.
    This includes:
    Games Played
    Games Won
    Games Lost
    Average Attempts per Game
    Best Round Score
    
    """

    if stats["games_played"] > 0:
        average_attempts = stats["total_attempts"] / stats["games_played"]
    else:
        average_attempts = 0

    print("\n" + "-" * 50)
    print("GAME STATISTICS".center(50))
    print("-" * 50)

    print(f"{'Games Played:':<30}{stats['games_played']}")
    print(f"{'Games Won:':<30}{stats['games_won']}")
    print(f"{'Games Lost:':<30}{stats['games_lost']}")
    print(f"{'Average Attempts per Game:':<30}{round(average_attempts, 2)}")
    print(f"{'Best Round Score:':<30}{stats['best_score']}")

    print("-" * 50 + "\n")
```

```python
# Main program

def main():
    """
    Main function controlling the overall flow of the game.
    It starts the session, runs the game loop, and displays
    final statistics when the player is done playing.
    """

    # Dictionary used to store statistics across all rounds in the session
    stats = {
        "games_played": 0,
        "games_won": 0,
        "games_lost": 0,
        "total_attempts": 0,
        "best_score": 0,
        "session_score": 0
    }
    
    # Ask the player for their name
    player_name = get_player_name()

    # Display the game instructions and scoring rules
    show_instructions(player_name)

    # Main game loop
    # The game continues running until the player decides to stop
    while True:

        # Run one round of the game
        play_game(player_name, stats)

        # Ask the player if they want to play another round
        # If the player confirms they want to stop, exit the loop
        if not ask_to_play_again():
            break

    # After the player stops playing, display the final session score
    print(f"Final session score: {stats['session_score']}")

    # Display overall game statistics for this session
    show_statistics(stats)

    # Inform the player that the session has ended
    print("\nSession ended.")
    print("Stats will reset next time you play.")
    print("Thanks for playing!\n")
```

```python
# Run the game

main()
```
